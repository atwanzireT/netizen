```python
df = pd.read_csv("/content/drive/MyDrive/coffee quality/arabica_data_cleaned.csv")
df
```

```python
df.dropna()
```

```python
dataset = df.loc[:, ["Species", "Country.of.Origin", "Aroma", "Flavor", "Moisture", ]]
```

```python
dataset
```

```python
dataset.describe()
```

```python
X = dataset.loc[:, ["Aroma", "Flavor"]]
```

```python
y = dataset.loc[:, ["Moisture"]]
```

##Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
X_test.shape
```

##Training Data

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```
